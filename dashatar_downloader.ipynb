```python
import os
import requests
import pandas as pd
```

```python
dashatars = pd.read_csv('dashatars.csv', index_col='role')
```

```python
def create_dir(path):
    if not os.path.exists(path):
        os.mkdir(path)
```

```python
folder = 'dashatars'

create_dir(folder)
for role in dashatars.index.unique():
    inner_folder = os.path.join(folder, role)
    create_dir(inner_folder)
    for i, row in dashatars.loc[role].iterrows():
        print('Downloading: %s: ' % row['image_url'], end='')
        response = requests.get(row['image_url'], allow_redirects=True)
        file_name = 's-%s_a-%s_w-%s_c-%s.png' % \
            (row['strength'], row['agility'], row['wisdom'], row['charisma'])
        
        with open(os.path.join(inner_folder, file_name), 'wb') as image_file:
            image_file.write(response.content)
        print('Done')
```
