```python
import json
import requests
import pandas as pd
```

```python
dashatars = {"manager": [], "designer": [], "developer": []}


def fetch_image(pair, role):
    queries = "agility=%s&wisdom=%s&strength=%s&charisma=%s&role=%s" % \
        (pair["agility"], pair["wisdom"], pair["strength"], pair["charisma"], role)

    response = requests.get("https://us-central1-dashatar-dev.cloudfunctions.net/createDashatar?" + queries)
    return json.loads(response.content)


def get_dashatar(i, j, k, l):
    if i + j + k + l != 12:
        return

    pair = {"strength": i, "agility": j, "wisdom": k, "charisma": l}
    for index, role in enumerate(["manager", "designer", "developer"]):
        pair["role"] = role
        image = fetch_image(pair, role)
        pair["image_url"] = image["url"]
        pair["image"] = image["fileName"]

        dashatars[role].append(dict(pair))
```

```python
for i in range(1, 6):
    for j in range(1, 6):
        for k in range(1, 6):
            for l in range(1, 6):
                get_dashatar(i, j, k, l)
```

```python
data = pd.DataFrame(columns=["role", "strength", "agility", "wisdom", "charisma", "image", "image_url"]);
```

```python
for role in dashatars:
    data = data.append(dashatars[role], ignore_index=True)
```
